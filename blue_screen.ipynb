# **Library to be get imported**

```
import matplotlib.pyplot as plt
import cv2
import numpy as np
```

# Reading the image using opencv


```
image = cv2.imread('/content/drive/MyDrive/image for computer vision/pizza_bluescreen.jpg')
image.shape
```

# Plotting or viewing the image

```
plt.imshow(image)
```

# The image was in the BGR format so we are converting it to the RGB format

```
image_as_rgb = cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
```

```
plt.imshow(image_as_rgb)
```

```
lower_pixel = np.array([0,0,220])
highest_pixel = np.array([50,50,255])
```

```
mask = cv2.inRange(image_as_rgb,lower_pixel,highest_pixel)
```

```
plt.imshow(mask, cmap='gray')
```

```
unmasked = np.copy(image_as_rgb)
unmasked[mask!=0] = [0,0,0]
```

```
plt.imshow(unmasked)
```

```
backround = cv2.imread('/content/drive/MyDrive/image for computer vision/space_background.jpg')
```

```

mask_backround = np.copy(backround)

mask_backround = backround[:514,:816]
mask_backround = cv2.cvtColor(mask_backround,cv2.COLOR_BGR2RGB)
mask_backround[mask==0] = [0,0,0]
backround_added = unmasked + mask_backround

plt.imshow(backround_added)
```

# when background is shady or with some light

```
shady_image_real = cv2.imread('/content/drive/MyDrive/image for computer vision/water_balloons.jpg')
```

```
plt.imshow(shady_image_real)
```

```
shady_image = cv2.cvtColor(shady_image_real,cv2.COLOR_BGR2RGB)
```

```
plt.imshow(shady_image)
```

```
hsv_image = cv2.cvtColor(shady_image_real,cv2.COLOR_BGR2HSV)
```

```
plt.imshow(hsv_image)
```

```
h = hsv_image[:,:,0]
s = hsv_image[:,:,1]
v = hsv_image[:,:,2]

f, (ax1,ax2,ax3) = plt.subplots(1,3,figsize=(20,10))
ax1.set_title('hue')
ax1.imshow(h, cmap='gray')

ax2.set_title('saturation')
ax2.imshow(s, cmap='gray')

ax3.set_title('value')
ax3.imshow(v, cmap='gray')
```

```
lower_rgb = np.array([180,0,100])
higher_rgb = np.array([255,255,230])
```

```
mask_rgb = cv2.inRange(shady_image, lower_rgb, higher_rgb)
```

```
plt.imshow(mask_rgb, cmap='gray')
```

```
lower_hsv = np.array([160,0,0])
higher_hsv = np.array([180,255,255])
```

```
mask_hsv = cv2.inRange(hsv_image,lower_hsv,higher_hsv)
```

```
image = np.copy(hsv_image)
image[mask_hsv==0] = [0,0,0]

plt.imshow(image, cmap='gray')
```
